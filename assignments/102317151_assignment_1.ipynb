```python
inputs = [(0,0),(0,1),(1,0),(1,1)]
```

```python
# AND Gate
print("AND Gate Output:")
for x1, x2 in inputs:
    s = x1*1 + x2*1
    if s >= 1.5:
        y = 1
    else:
        y = 0
    print(x1, x2, " ", y)
```

For AND gate, both inputs need to be 1 so threshold is kept at 1.5 with weights w1=1, w2=1.

```python
# OR Gate
print("OR Gate Output:")
for x1, x2 in inputs:
    s = x1*1 + x2*1
    if s >= 0.5:
        y = 1
    else:
        y = 0
    print(x1, x2, " ", y)
```

For OR gate, even one input being 1 is enough, so threshold is lowered to 0.5.

```python
# NAND Gate
print("NAND Gate Output:")
for x1, x2 in inputs:
    s = x1*(-1) + x2*(-1)
    if s >= -1.5:
        y = 1
    else:
        y = 0
    print(x1, x2, " ", y)
```

NAND is opposite of AND. Used negative weights -1, -1 and threshold -1.5 to invert the AND behavior.

```python
# NOR Gate
print("NOR Gate Output:")
for x1, x2 in inputs:
    s = x1*(-1) + x2*(-1)
    if s >= -0.5:
        y = 1
    else:
        y = 0
    print(x1, x2, " ", y)
```

NOR is opposite of OR. Same negative weights but threshold changed to -0.5 to invert OR behavior.

```python
# General Perceptron function
def perceptron(x1, x2, w1, w2, bias):
    net = x1 * w1 + x2 * w2
    return 1 if net >= 0 else 0
```

```python
# Testing perceptron for all gates
print("Testing AND using perceptron:")
for x1, x2 in inputs:
    print(x1, x2, "->", perceptron(x1, x2, 1, 1, -1.5))
```
