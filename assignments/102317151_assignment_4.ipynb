```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("multiple_linear_regression_dataset.csv")

print(data.head())
print(data.shape)
```

```python
# using age and experience as features to predict income
X = data[["age", "experience"]]
y = data["income"]
```

```python
# initializing weights and bias to zero
n_features = X.shape[1]

w = np.zeros(n_features)
b = 0.0
```

```python
def predict(X, w, b):
    return X.dot(w) + b

def mean_squared_error(y, y_hat):
    return ((y_hat - y) ** 2).mean()

def compute_gradients(X, y, y_hat):
    N = len(y)
    dw = (2 / N) * X.T.dot(y_hat - y)
    db = (2 / N) * (y_hat - y).sum()
    return dw, db

def update_parameters(w, b, dw, db, lr):
    w = w - lr * dw
    b = b - lr * db
    return w, b
```

Training the model using gradient descent. Learning rate is kept small (0.0001) to avoid overshooting.

```python
lr = 0.0001
epochs = 1000

for epoch in range(epochs):
    y_hat = predict(X, w, b)
    loss = mean_squared_error(y, y_hat)
    dw, db = compute_gradients(X, y, y_hat)
    w, b = update_parameters(w, b, dw, db, lr)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")
```

```python
print("Final Weights:", w)
print("Final Bias:", b)

# predicting income for a 35 year old with 6 years experience
new_person = np.array([35, 6])
predicted_income = new_person.dot(w) + b

print("Predicted Income:", predicted_income)
```

The model learned that experience has a higher weight (~1371) compared to age (~764), meaning experience contributes more to income prediction. For a 35 year old with 6 years of experience, the predicted income comes out to around 35314 which seems reasonable based on the training data range.
