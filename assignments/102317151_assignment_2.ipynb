```python
import numpy as np
import pandas as pd
```

```python
# generating random data for glass samples
np.random.seed(42)

sodium = np.random.uniform(12, 18, 6)
magnesium = np.random.uniform(1.5, 2.8, 6)
calcium = np.random.uniform(7, 11.5, 6)
ri = np.random.uniform(1.50, 1.55, 6)
```

```python
# putting everything into a dataframe
df = pd.DataFrame({
    "Sodium": sodium,
    "Magnesium": magnesium,
    "Calcium": calcium,
    "Refractive_Index": ri
})
```

Setting thresholds for Sodium and Calcium. If either exceeds the limit, sample is rejected.

```python
SODIUM_THRESHOLD = 15.5
CALCIUM_THRESHOLD = 10.0

# if sodium > 15.5 OR calcium > 10 then reject, else accept
df["Decision"] = np.where(
    (df["Sodium"] > SODIUM_THRESHOLD) | (df["Calcium"] > CALCIUM_THRESHOLD),
    "Reject",
    "Accept"
)
```

```python
print(df)
```

Out of 6 samples, 4 got rejected and 2 were accepted. Sample 4 and 5 passed both thresholds so they are accepted. This is basically working like a perceptron where the decision boundary separates acceptable and non-acceptable glass based on chemical composition.
